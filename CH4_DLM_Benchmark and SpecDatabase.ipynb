```python
import matplotlib.pyplot as plt
from radis import plot_diff, SpectrumFactory, get_residual
from publib import set_style, fix_style
```

```python
from radis.io.hitemp import fetch_hitemp

fetch_hitemp('CH4')
```

**Setting Parameters**

```python
#%% Computation parameters
# List of min and max wavenumber
wmin = [2900, 2900, 2000, 2000]
wmax = [3000, 4000, 4000, 4500]
dv = 0.002
T = 3000.0 #K
p = 0.1 #bar
```

**Creating Spectrum objects for different spectral range**

```python
from radis import calc_spectrum
spectra_DLM = {}
N_lines = []

for i, j in zip(wmin,wmax):
  s = calc_spectrum(i, j,         # cm-1
                  molecule='CH4',
                  pressure=p,   # bar
                  Tgas=T,
                  databank='hitemp',  # or 'hitemp'
                  optimization='simple',
                  verbose = 3,
                  wstep = dv,
                  )
  s.name = 'DLM simple ({0:.1f}s)'.format(s.conditions['calculation_time'])
  N = s.conditions['total_lines']
  N_lines.append(N)
  spectra_DLM[N] = s
```

```python

#%% Compare performance for N_lines
set_style('origin')
plt.figure()
plt.plot(N_lines, [spectra_DLM[N].conditions['calculation_time'] for N in N_lines], '--ok', label = "Optimization = simple" )
plt.xlabel('Number of lines (#)')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Calculation time (s)')
plt.legend()
fix_style('origin')
```

# Saving spectrum using store()

```python
spectra_DLM[N_lines[0]].store('/root/test.spec')
```

**Initializing Spec Database and Saving Spectrum through it**

```python
from radis.tools import SpecDatabase
db = SpecDatabase(r"/CH4/")   
```

```python
db.add(spectrum=spectra_DLM[N_lines[0]],store_name='/CH4/test.spec')
db.add(spectrum=spectra_DLM[N_lines[1]],store_name='/CH4/test1.spec')
```
